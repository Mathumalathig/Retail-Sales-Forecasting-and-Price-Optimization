```python
import numpy as np
import pandas as pd
```

```python
## Load & Merge Datasets
```

```python
train = pd.read_csv("../data/train.csv")
feature = pd.read_csv("../data/features.csv")
stores = pd.read_csv("../data/stores.csv")
```

```python
df = train.merge(feature, how = 'left', on =('Store','Date'))
df = df.merge(stores, how = 'left', on =('Store'))
df.head()
```

```python
## Convert to date
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
missing_values = df.isnull()
```

```python
missing_values.sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df['Month'] = df['Date'].dt.month
df['Week'] = df['Date'].dt.isocalendar().week
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
## Weekly Seasonality
```

```python
df.groupby('Week')['Weekly_Sales'].mean().plot()
plt.title("Average Weekly Sales(Sesonality)")
plt.xlabel("Week Number")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python
## Top 10 performing Stores
```

```python
top_stores = df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False).head(10)
top_stores.plot(kind='bar', color='skyblue')
plt.title('Top 10 performing Stores')
plt.xlabel("Stores")
plt.ylabel("Total Sales")
plt.show()
```

```python
## Sales trends over time
```

```python
df.groupby('Date')['Weekly_Sales'].sum().plot(figsize=(12,6))
plt.title("Sales Trend over time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

```python
df_store = df[(df['Store']==1) & (df['Dept']==1)].copy()
df_store.count()
```

```python
df_store = df_store.sort_values('Date')
df_store.head()
```

```python
sales_week = df_store.groupby('Date')['Weekly_Sales'].sum()
print(sales_week.head(10))
print(sales_week.tail(10))
print(sales_week.describe())
```

```python
sales_week = df_store.set_index('Date').resample('W')['Weekly_Sales'].sum()
```

```python
sales_week.isnull().sum()
```

```python
## visvalize the weekly_sales
```

```python
sales_week.plot(figsize=(12,6), title = "Weekly_sales for store 1 and dept 1")
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomp = seasonal_decompose(sales_week, model='additive', period=12)
decomp.plot()
```

```python
train = sales_week[:-10]
test = sales_week[:10]
```

```python
## Naive vs Actual
```

```python
naive_forecast = train.iloc[-1]
naive_predict = [naive_forecast] * len(test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
naive_mae = mean_absolute_error(test, naive_predict)
naive_rmse = np.sqrt(mean_squared_error(test,naive_predict))

print("MAE:", round(naive_mae, 2))
print("RMSE:", round(naive_rmse, 2))
```

```python
plt.figure(figsize=(12,6))
plt.plot(train.index,train,label='Train')
plt.plot(test.index,test,label='Actual')
plt.plot(test.index,naive_predict,label='Naive Forecast')
plt.title('Naive vs Actual')
plt.legend()
plt.show()
```

```python
## Moving Average Forecast vs Actual
```

```python
window_size = 3
move_avg_forecast = train.rolling(window=window_size).mean().iloc[-1]
move_avg_predict = [move_avg_forecast] * len(test)
```

```python
mae = mean_absolute_error(test, naive_predict)
rmse = np.sqrt(mean_squared_error(test,naive_predict))
```

```python
print("MAE:", round(mae, 2))
print("RMSE:", round(rmse, 2))
```

```python
plt.figure(figsize=(12,6))
plt.plot(train.index,train,label='Train')
plt.plot(test.index,test,label='Actual')
plt.plot(test.index,move_avg_predict,label='Moving Avg Forecast',linestyle='--')
plt.title('Moving Average Forecast vs Actual')
plt.legend()
plt.show()
```

```python
## ARIMA Forecasting
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(train)
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train, order=(1,1,1))
model_fit = model.fit()

print(model_fit.summary())
```

```python
forecast = model_fit.forecast(steps=len(test))
```

```python
arima_mae = mean_absolute_error(test,forecast)
arima_rmse = np.sqrt(mean_squared_error(test,forecast))
```

```python
print('MAE:',round(arima_mae,2))
print('RMSE:',round(arima_rmse,2))
```

```python
plt.figure(figsize=(12,6))
plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Actual')
plt.plot(test.index, forecast, label='ARIMA Forecast', linestyle='--')
plt.title("Arima Forecast vs Actual")
plt.legend()
plt.show()
```

```python
## Model Comparison summary 
```

```python
results = pd.DataFrame({
    'Model' : ['Naive Forecast', 'ARIMA'],
    'MAE' : [naive_mae, arima_mae],
    'RMSE' : [naive_rmse, arima_rmse]
})

print(results)
```

```python
plt.figure(figsize=(12,6))
plt.plot(train.index, train, label="Train")
plt.plot(test.index, test, label="Actual")
plt.plot(test.index, naive_predict, label="Naive Forecast", linestyle='--', color='green')
plt.plot(test.index, forecast, label='Arima Forecast', linestyle='--', color='red')
plt.title("Model Comaparison : Naive vs ARIMA")
plt.legend()
plt.show()
```
